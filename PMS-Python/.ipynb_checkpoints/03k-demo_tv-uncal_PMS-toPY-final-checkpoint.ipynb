```python
import matplotlib.pyplot as plt

norms_fake_na_lst = []
norms_fake_lst = []
norms_pms_lst = []
bot_lst = []

for bot in range(96,106):
    norms_fake_na = np.load(f'./bottle_info/{bot}_norms_fake_na_.npy')
    norms_fake = np.load(f'./bottle_info/{bot}_norms_fake_.npy')
    norms_pms = np.load(f'./bottle_info/{bot}_norms_pms_.npy')
    
    bot_lst.append(bot)
    norms_fake_na_lst.append(norms_fake_na.mean())
    norms_fake_lst.append(norms_fake.mean())
    norms_pms_lst.append(norms_pms.mean())

norms = pd.DataFrame([bot_lst, norms_fake_na_lst, norms_fake_lst, norms_pms_lst], index=['bot_num',
                                                                                 'norms_fake_na_lst',
                                                                                 'norms_fake_lst',
                                                                                 'norms_pms_lst']).T
norms['bot_num'] = norms['bot_num'] - 95

norms['set'] = None
norms.loc[norms['bot_num'] <= 100, 'set'] = 'valid'
norms.loc[norms['bot_num'] > 100, 'set'] = 'test'
norms = norms.set_index('bot_num')
norms
```

```python
plt.figure(figsize=(7,6))
width = 0.2
plt.bar(x=norms.index-width,height=norms['norms_fake_na_lst'],width=width)
plt.bar(x=norms.index,height=norms['norms_fake_lst'],width=width)
plt.bar(x=norms.index+width,height=norms['norms_pms_lst'],width=width)
plt.grid()
plt.xticks(np.array(bot_lst)-95)
plt.xlabel('Bottle Number (1-5 Validation, 6-10 Test)')
plt.ylabel('Avergage Euclidean Distances from True Point Cloud')
plt.title('Euclidean Distances vs. Bottle Number by Methodology')
plt.legend(['Fake — No Avg.','Fake — Avg.','Pure PMS'], loc='best')
plt.savefig('PMS_norm_err_fig1.jpg',dpi=300)
```

```python
import cv2
def cvt_to_dims(img_num):
    fn = f'./test_rand/fake/{img_num}.jpg'
    
    tmp = np.array(cv2.imread(fn, cv2.IMREAD_GRAYSCALE)).astype(float)
    tmp = np.pad(tmp, [((128 - tmp.shape[0])//2,(128 - tmp.shape[0])//2), ((128 - tmp.shape[1])//2,(128 - tmp.shape[1])//2)])
    tmp = (tmp >= 0) * tmp * 255 / tmp.max()
    tmp = np.round(tmp, 0)
    tmp = tmp.astype(float)
    
    return tmp

for img_num in range(1,7):
    tmp_img = np.array(cvt_to_dims(img_num))
    plt.imshow(tmp_img)
    plt.show()
    cv2.imwrite(f'./padded_can_{img_num}.jpg',tmp_img)
```
